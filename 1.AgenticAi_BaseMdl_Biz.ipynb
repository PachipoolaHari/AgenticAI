```python
# 
from openai import OpenAI
```

```python
#This is the working code for Gemini. 
#Note the model i used here is gemini-2.0-flash

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
```

```python
# And now - using the model appropriate for the API call to end point?

response = gemini.chat.completions.create(model="gemini-2.0-flash", messages=[{"role":"user", "content": "what is 2+2?"}])
```

```python
# Ask it again for resource allocation and prioritization conflicts in agile teams

question = "Propose an Agentic AI solution for the pain point on, Resource Allocation & Prioritization Conflicts in Agile Teams ."
messages=[{"role":"user", "content":question}]

response=gemini.chat.completions.create(
    model="gemini-2.0-flash",
    messages=messages
    )


answer = response.choices[0].message.content
print(answer)
```
